```python
import pandas as pd
```

```python
dataframe = pd.read_csv("COVID-19 Global Statistics Dataset.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
dataframe.head()
```

```python
dataframe['Total Cases'] = dataframe['Total Cases'].str.replace(',','').astype(float)
```

```python
dataframe['Total Cases'] = dataframe['Total Cases'].fillna(0)
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(x='Country', y='TotalCases', dataframe=dataframe.nlargest(10, 'Total Cases'))
plt.title('Top 10 Countries by Total Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```
