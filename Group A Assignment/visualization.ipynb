```python
import pandas as pd
```

```python
dataframe = pd.read_csv("student_dataset.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
age_sum = dataframe.groupby('name')['age'].sum()
```

```python
age_sum.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Age Distribution by Name')
plt.ylabel('')  # This removes the y-label
plt.show()
```

```python
total_cases_for_countries = dataframe.groupby('Country')['Total Cases'].sum().sort_values(ascending=False)
```

```python
top_10_countries = total_cases_for_countries.head(20) 
```

```python
print(top_10_countries[1:-1])
```

```python
top_10 = top_10_countries[1:-1]
```

```python
print(top_10)
```

```python
plt.figure(figsize=(14, 8))
top_10.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries by Total Cases')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.xticks(rotation=45)
plt.show()
```
