```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv('student_dataset.csv')
```

```python
dataframe['ages'] = pd.cut(dataframe['Age'], bins=[17, 23, 28, 33, 38],labels=['17-23', '24-28', '29-33', '34-38'])
```

```python
age_sum = dataframe['ages'].value_counts()
```

```python
plt.pie(age_sum, labels=age_sum.index, autopct='%1.1f%%', startangle=(140))
plt.axis('equal')
plt.title('Age Distribution by Name')
plt.show()
```
