```python
import pandas as pd
```

```python
data = pd.read_csv("COVID-19 Global Statistics Dataset.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
data.head()
```

```python
columns_to_process = ['Total Deaths', 'Total Recovered', 'Active Cases']
```

```python
for column in columns_to_process:
    data[column] = data[column].str.replace(',','').astype(float)
    data[column] = data[column].fillna(0)
```

```python
data.head(20)
```

```python
total_deaths = data['Total Deaths'].sum()
total_recovered = data['Total Recovered'].sum()
active_cases = data['Active Cases'].sum()
```

```python
print('Total deaths:', total_deaths)
print('Total recovered:', total_recovered)
print('Active cases:', active_cases)
```
